```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

df = pd.read_csv(r"C:\Users\dines\Downloads\Business_Analytics\clean_airline_dataset.csv")
# Count distribution
gender_counts = df["Gender"].value_counts()

print("Gender Distribution:")
print(gender_counts)

# Percentage
gender_percent = (gender_counts / len(df)) * 100
print("\nPercentage Distribution:")
print(gender_percent.round(2))
```

```python
plt.figure()
sns.countplot(data=df, x="Gender")
plt.title("Gender Distribution of Airline Customers")
plt.xlabel("Gender")
plt.ylabel("Number of Customers")
plt.show()

plt.figure()
ax = sns.countplot(data=df, x="Gender")

total = len(df)
for p in ax.patches:
    percentage = f'{100 * p.get_height() / total:.1f}%'
    ax.annotate(percentage,
                (p.get_x() + p.get_width() / 2., p.get_height()),
                ha='center', va='bottom')

plt.title("Gender Distribution (Percentage)")
plt.show()
```

```python
import ast

def parse_influencing(val):
    if pd.isna(val):
        return []
    
    val = str(val).strip()
    
    if val.startswith("[") and val.endswith("]"):
        try:
            parsed = ast.literal_eval(val)
            return [str(x).strip().lower() for x in parsed]
        except:
            return []
    
    return [val.lower()]


df["Influencing_List"] = df["Influencing_Factors"].apply(parse_influencing)

influence_exploded = df.explode("Influencing_List")

influence_exploded = influence_exploded[
    influence_exploded["Influencing_List"] != ""
]

influence_counts = influence_exploded["Influencing_List"].value_counts()

print(influence_counts)
```

```python
influence_airline_percent = pd.crosstab(
    influence_exploded["Influencing_List"],
    influence_exploded["Airline_Selected"],
    normalize="columns"
) * 100

print(influence_airline_percent.round(1))
```
